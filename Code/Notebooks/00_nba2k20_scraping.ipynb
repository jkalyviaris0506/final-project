### scrape the data 
Data is coming from : https://www.basketball-reference.com/leagues/NBA_2020_per_game.html

From the website we are scraping the players names, age, team, position, along with their game statistics


```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By # used to import different ways to access data in the XML or HTML file
from selenium.webdriver.chrome.service import Service # no longer need to download a driver file, use service
from webdriver_manager.chrome import ChromeDriverManager # used to manage the Chrome driver to emulate a Chrome web browser
import time
import random
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
# function to scroll from the top to the bottom of the web page
def random_scroll(browser, total_wait_time):
    # get the total height of the page
    total_height = browser.execute_script("return document.body.scrollHeight")
    
    # number of steps to scroll (you can adjust this number)
    scroll_steps = random.randint(3, 10) # randomize how many scroll steps we will use
    
    # calculate the height to scroll on each step
    scroll_increment = total_height // scroll_steps

    # calculate the total time available for scrolling each step
    time_per_step = total_wait_time / scroll_steps
    
    # random scrolling across time
    for step in range(scroll_steps):
        # scroll by the increment (dividing total height by number of steps)
        browser.execute_script(f"window.scrollBy(0, {scroll_increment});")
        
        # random wait time between scrolls to simulate varying speed
        random_wait = random.uniform(0.5 * time_per_step, 1.5 * time_per_step)  # randomize the wait within a range
        time.sleep(random_wait)
        
    # final scroll to make sure you are at the very bottom (in case it didn't exactly match)
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")

# initialize the Selenium web driver (Chrome in this case)
browser = webdriver.Chrome()

# create empty lists to store the extracted data
player_name = []
player_age = []
team_name = []
position = []
ppg=[]
assists_per_game=[]
rebounds_per_game=[]
min_played=[]


#for i in range(1, ): # first value is inclusive, second value is not inclusive
url = f"https://www.basketball-reference.com/leagues/NBA_2020_per_game.html"

print(url)

    # navigate to the web page using the URL
browser.get(url)
browser.maximize_window()

    # add a random delay before scraping
total_wait_time = random.uniform(2, 20)  # random wait time between 3 and 10 seconds
random_scroll(browser, total_wait_time)

    # scraping the individual elements from the page
player_elements = browser.find_elements(By.XPATH, "//table[@id='per_game_stats']//tbody//tr//td[@data-stat='name_display']")
age_elements= browser.find_elements(By.XPATH, "//table[@id='per_game_stats']//tbody//tr//td[@data-stat='age']")
team_elements= browser.find_elements(By.XPATH, "//table[@id='per_game_stats']//tbody//tr//td[@data-stat='team_name_abbr']")
position_elements= browser.find_elements(By.XPATH, "//table[@id='per_game_stats']//tbody//tr//td[@data-stat='pos']")
ppg_elements= browser.find_elements(By.XPATH, "//table[@id='per_game_stats']//tbody//tr//td[@data-stat='pts_per_g']")
assists_elements= browser.find_elements(By.XPATH, "//table[@id='per_game_stats']//tbody//tr//td[@data-stat='ast_per_g']")
rebound_elements= browser.find_elements(By.XPATH, "//table[@id='per_game_stats']//tbody//tr//td[@data-stat='trb_per_g']")
min_elements= browser.find_elements(By.XPATH, "//table[@id='per_game_stats']//tbody//tr//td[@data-stat='mp_per_g']")



for name_display in player_elements:
    player_name.append(name_display.text)

for age in age_elements:
    player_age.append(age.text)

for team_name_abbr in team_elements:
    team_name.append(team_name_abbr.text)

for pos in position_elements:
    position.append(pos.text)

for pts_per_g in ppg_elements:
    ppg.append(pts_per_g.text)

for ast_per_g in assists_elements:
    assists_per_game.append(ast_per_g.text)

for trb_per_g in rebound_elements:
    rebounds_per_game.append(trb_per_g.text)

for mp_per_g in min_elements:
    min_played.append(mp_per_g.text)

# close the automated browser
browser.close()

# report out scraped data stats
print("-"* 30)
print(f"# of players: {len(player_name)}")
```

```python
original_nba_stats= pd.DataFrame({
    'player':player_name,
    'age':player_age,
    'team':team_name,
    'position':position,
    'ppg':ppg,
    'assists_per_game':assists_per_game,
    'rebounds_per_game':rebounds_per_game,
    'min_per_game':min_played
    
    
    
})
```
