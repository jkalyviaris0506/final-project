```python
# load in nba2k20 data 
import pandas as pd 
```

```python
nba2k20_raw= pd.read_csv('nba2k-full.csv')
```

```python
nba2k20_raw
```

```python
nba2k20_edit= nba2k20_raw.copy()
```

```python
nba2k20_edit=nba2k20_edit.drop(columns='b_day')
```

```python
nba2k20_edit=nba2k20_edit.drop(columns='country')
```

```python
nba2k20_edit=nba2k20_edit.drop(columns='draft_year')
```

```python
nba2k20_edit=nba2k20_edit.drop(columns='draft_round')
```

```python
nba2k20_edit=nba2k20_edit.drop(columns='draft_peak')
```

```python
nba2k20_edit=nba2k20_edit.drop(columns='college')
```

```python
nba2k20_edit=nba2k20_edit.drop(columns='version')
```

```python
nba2k20_edit=nba2k20_edit.drop(columns='jersey')
```

```python
nba2k20_edit.isna().sum()
```

```python
nba2k20_edit['height']=nba2k20_edit['height'].str.split('/').str[0]
```

```python
nba2k20_edit['weight']=nba2k20_edit['weight'].str.split('/').str[0]
```

```python
nba2k20_edit['weight']=nba2k20_edit['weight'].str.split('lbs.').str[0]
```

```python
nba2k20_edit.head()
```

```python
def convert_height_to_inches(height_str):
    feet, inches = map(int, height_str.split('-'))
    return feet * 12 + inches

# Apply the function to the 'Height' column
nba2k20_edit['height'] = nba2k20_edit['height'].apply(convert_height_to_inches)
```

```python
nba2k20_edit
```

```python
nba2k20_edit.to_csv('nba2k20_data_edited.csv')
```
